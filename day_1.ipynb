## Practice Questions


1. How many students are from the electronics Department ?
2. What is the average attendence of Female Students ?
3. Who is the student with lowest programming score ?
4. How many students scored above 80 in math ?
5. What is the average total_score for students with attendence above 90% ?

```python
# Total students from Electronics department

total_ECE_stud = df['department'].value_counts()['Electronics']

print(f"Number of students in the Electronics Department: {total_ECE_stud}")
```

```python
# Average Attendence of Female Students

Avg_fem_stud=df[df['gender']=='Female']['attendance_percentage'].mean().round(2)
print("Average Attendence of Female Students ",Avg_fem_stud)
```

```python
#Student with Lowest Programming Score

low_prog_score = df.loc[df['programming_score'].idxmin()]['name']
print("Student with lowest Programming score : ",low_prog_score)
```

```python
# Students scored 80 above on math

math_score_above80 = df[df['math_score']>80]['name'].count()

math_score_above80
```

```python
# Average total score of srudents with attendence above 90 %

avg_total_score = df[df['attendance_percentage']>90]['total_score'].mean().round(2)

avg_total_score
```
